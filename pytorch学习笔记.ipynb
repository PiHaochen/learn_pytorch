## 十、随机梯度下降

### 1、torch.autograd.grad(loss,[w1,w2,...])

```python
from torch.nn import functional as F
```

```python
x=torch.ones(1)
w=torch.full([1],2)
mse=F.mse_loss(torch.ones(1),x*w)
```

```python
torch.autograd.grad(mse,[w])
```

```python
w.requires_grad_()#把w标记为需要求导
```

```python
torch.autograd.grad(mse,[w])
```

```python
mse=F.mse_loss(torch.ones(1),x*w)#重新生成计算图
```

```python
torch.autograd.grad(mse,[w])
```

### 2、loss.backword

```python
mse=F.mse_loss(torch.ones(1),x*w)
```

```python
mse.backward()
```

```python
w.grad
```

### 3、softmax激活函数

```python
a=torch.rand(3)
a.requires_grad_()
```

```python
p=F.softmax(a,dim=0)
```

```python
p[1].backward()#法一
a.grad
```

```python
p=F.softmax(a,dim=0)
torch.autograd.grad(p[1],[a],retain_graph=True)#法二
```

```python
torch.autograd.grad(p[2],[a])
```
